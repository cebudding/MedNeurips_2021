```python
def flatten_ground_truth(ground_truth, len_set): 
    ground_truths = []
    for idx in range(len_set): 
        slice_idx = idx % 135
        part_idx = int(idx/135)
        ground_truths.append(ground_truth[part_idx][slice_idx][2])
    return ground_truths 
```

## Generate perlin noise with brain mask

```python
def generate_perlin_noise_2d(shape, res, **kwargs):
    """
    Adapted from https://pvigier.github.io/2018/06/13/perlin-noise-numpy.html
    
    This function generates 2D perlin noise that can be used as a background for artificial 
    lesion data. 
    
    Parameters: 
    shape : shape of the image 
    res : TODO figure out what this does exactly 
    
    """
    def f(t):
        return 6*t**5 - 15*t**4 + 10*t**3
    
    delta = (res[0] / shape[0], res[1] / shape[1])
    d = (shape[0] // res[0], shape[1] // res[1])
    grid = np.mgrid[0:res[0]:delta[0],0:res[1]:delta[1]].transpose(1, 2, 0) % 1
    # Gradients
    angles = 2*np.pi*np.random.rand(res[0]+1, res[1]+1)
    gradients = np.dstack((np.cos(angles), np.sin(angles)))
    g00 = gradients[0:-1,0:-1].repeat(d[0], 0).repeat(d[1], 1)
    g10 = gradients[1:,0:-1].repeat(d[0], 0).repeat(d[1], 1)
    g01 = gradients[0:-1,1:].repeat(d[0], 0).repeat(d[1], 1)
    g11 = gradients[1:,1:].repeat(d[0], 0).repeat(d[1], 1)
    # Ramps
    n00 = np.sum(grid * g00, 2)
    n10 = np.sum(np.dstack((grid[:,:,0]-1, grid[:,:,1])) * g10, 2)
    n01 = np.sum(np.dstack((grid[:,:,0], grid[:,:,1]-1)) * g01, 2)
    n11 = np.sum(np.dstack((grid[:,:,0]-1, grid[:,:,1]-1)) * g11, 2)
    # Interpolation
    t = f(grid)
    n0 = n00*(1-t[:,:,0]) + t[:,:,0]*n10
    n1 = n01*(1-t[:,:,0]) + t[:,:,0]*n11
    return np.sqrt(2)*((1-t[:,:,1])*n0 + t[:,:,1]*n1)
```

```python
def get_perlin_data_brain_lesions(brain_data, ground_truth_data, shape, res, feature_range = (0, 1), blur = True): 
    data = []
    mri_data = []
    for i in range(len(brain_data)): 
        background_img = generate_perlin_noise_2d(shape = shape, res = res)
        background_img = minmax_scale(background_img.flatten(), feature_range = (0, 1)).reshape(background_img.shape)
        brain_mask = brain_data[i] > 0 
        background_img *= brain_mask 
#         data += [background_img]
        
        ground_truth_mask = ground_truth_data[i]
        background_img *= (ground_truth_mask * -1) + 1
        
        if blur: 
            background_img = gaussian_filter(background_img, sigma = 3)
            mri_img_blur = gaussian_filter(brain_data[i], sigma = 3)

        data += [background_img]
        mri_data += [mri_img_blur]
        
    return np.array(data), np.array(mri_data)   
```

```python
X_training_perlin, X_training_mri = get_perlin_data_brain_lesions(X_train, ground_truth_train, shape = (140, 192), res = res)
```

```python
X_holdout_perlin, X_holdout_mri = get_perlin_data_brain_lesions(X_holdout, ground_truth_holdout, shape = (140, 192), res = res)
```

```python
def save_data_pkl(array, data_dir, file_name):
    with open(os.path.join(data_dir, f'{file_name}.pkl'), 'wb') as f: 
        pkl.dump(array, f)
        
def save_data_h5py(list_data, list_labels, data_dir, file_name): 
    all_data = list_data
    all_labels = list_labels
    assert len(all_data) == len(all_labels)
    
    h5 = h5py.File(os.path.join(data_dir, f'{file_name}.h5'), 'w')
    h5.create_dataset('X', data=all_data, compression='gzip', compression_opts=9)
    h5.create_dataset('y', data=all_labels, compression='gzip', compression_opts=9)
    h5.close() 
```
